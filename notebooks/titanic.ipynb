```python
dataset = pd.read_csv('titanic_data.csv')
```

```python
dataset.head()
```

```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

# Convert categorical features to numerical using Label Encoding
le = LabelEncoder()
for column in ['Sex', 'Embarked', 'Cabin']:  # Replace with actual categorical columns
    if column in dataset.columns:
        dataset[column] = le.fit_transform(dataset[column].astype(str))
```

```python
# Handle missing values (replace with mean or median for numerical, mode for categorical)
dataset['Age'].fillna(dataset['Age'].mean(), inplace=True)
# Replace NaN values in Embarked with the mode (most frequent value)
dataset['Embarked'].fillna(dataset['Embarked'].mode()[0], inplace=True)
```

```python

# Drop unnecessary columns
dataset = dataset.drop(['Name', 'Ticket', 'PassengerId'], axis=1)
```

```python

# Remove Outliers using IQR
for column in ['Age', 'Fare']:  # Replace with columns you want to remove outliers
    Q1 = dataset[column].quantile(0.25)
    Q3 = dataset[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    dataset = dataset[(dataset[column] >= lower_bound) & (dataset[column] <= upper_bound)]
```

```python
# Separate features (X) and target (y)
X = dataset.drop('Survived', axis=1)
y = dataset['Survived']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
# Scale numerical features using StandardScaler
scaler = StandardScaler()
numerical_features = ['Age', 'Fare', 'Pclass', 'SibSp', 'Parch']  # Replace with your numerical features
X_train[numerical_features] = scaler.fit_transform(X_train[numerical_features])
X_test[numerical_features] = scaler.transform(X_test[numerical_features])
```

```python
# prompt: Build an randmforast model with X_train and validate on X_tet

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Create a Random Forest Classifier
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
rf_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf_model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```
